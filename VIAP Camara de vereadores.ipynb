```python
# IMPORTA BIBLIOTECAS E APIS


from selenium import webdriver
from selenium.webdriver.support.ui import Select
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.desired_capabilities import DesiredCapabilities

from datetime import date
import re
import pandas as pd
from unidecode import unidecode

#Não esperar o carregamento completo
caps = DesiredCapabilities().CHROME
caps["pageLoadStrategy"] = "normal"  #  interactive

# Dados para o selenium funcionar no windows
options = Options()
options.headless = False
options.add_argument('--no-sandbox')  # Bypass OS security model
options.add_argument('--disable-gpu')  # applicable to windows os only
#options.add_argument('start-maximized')  #
options.add_argument('disable-infobars')
options.add_argument("--disable-extensions")
options.add_argument('--disable-dev-shm-usage')
options.add_argument('window-size=1920x1480')

driver = webdriver.Chrome(desired_capabilities=caps, options=options)
```

```python
listalinks = []
i = 1

while i <= 7:
    driver.get("https://www.maceio.al.leg.br/transparencia/portal/viap&pagina=" + str(i))
    links = driver.find_elements_by_xpath("//*[@title='Visualizar Detalhes']")
    for link in links:
        listalinks.append(link.get_attribute('href'))
        print(link.get_attribute('href'))
    i = i +1
```

```python
listacompleta = []

dicionario = {"vereador":"", "mes":"", "ano":"", "saldomesanterior":"", "gastosdomes":"", "despesasglosadas":"", "excessodegasto":"", "saldoindenizado":"" }
```

```python
for link in listalinks:
    driver.get(link)

    #Entra na tabela
    for tabela in driver.find_elements_by_tag_name("tbody"):
        linhas = tabela.find_elements_by_tag_name("tr")

        for linha in linhas:
            time.sleep(0.3)
            mes = linha.text.split(" ")[0]
            ano = linha.text.split(" ")[2].split('\n')[0]
            saldomesanterior =linha.text.split(" ")[4]
            gastosdomes = linha.text.split(" ")[5]
            despesasglosadas = linha.text.split(" ")[6]
            excessodegasto = linha.text.split(" ")[7]
            saldoindenizado = linha.text.split(" ")[8]
            vereador = nomeano = driver.find_element_by_xpath("//*[@class='h2-responsive green-text font-weight-bold']").text.split('- ')[1].split(' / ')[0]
            
            
            #rint(vereador, mes, ano, saldomesanterior, gastosdomes, despesasglosadas, excessodegasto, saldoindenizado)
            dicionario.update({"vereador":vereador, "mes":mes, "ano":ano, "saldomesanterior":saldomesanterior, "gastosdomes":gastosdomes, "despesasglosadas":despesasglosadas, "excessodegasto":excessodegasto, "saldoindenizado":saldoindenizado })
            listacompleta.append(dicionario.copy())
            print(dicionario)
            print("--")
            

        
    
```

```python
df = pd.DataFrame(listacompleta)
```
